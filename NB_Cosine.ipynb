````python
import pandas as pd
import numpy as np

ResNet = pd.read_csv('./data/features_train/features_resnet1000_train.csv', header=None)
ResNet.columns = ['fnum'] + list(range(1000))
ResNet['fnum'] = ResNet['fnum'].apply(lambda x: int(x.split('/')[-1].split('.')[0]))
ResNet.sort_values('fnum', inplace = True)
ResNet.set_index('fnum', inplace = True)

# Manual implementation of Softmax
# Matrix of probabilities

# img num 1: [ P(class=1), P(class=2), ...]
# img num 2: [ P(class=1), P(class=2), ...]
probabilities = np.exp(ResNet.values)/np.exp(ResNet.values).sum(axis=1, keepdims=True)
probabilities

# image_class = np.argmax(probabilities)
image_class = np.argmax(probabilities, axis=1)
````

````python
token_vectors = pd.read_csv('tokens.csv')
token_vectors.set_index('Unnamed: 0', inplace=True)
test_token_vectors = pd.read_csv("./tokens_test.csv")
test_token_vectors.set_index('Unnamed: 0', inplace=True)
````

````python
from sklearn.naive_bayes import MultinomialNB 
classifiers = list()

for i in range(1000):
    print("Training model for ResNet101 Class: ",i)
    trained_model = MultinomialNB().fit(token_vectors.values, (image_class == i).astype(int))
    print((image_class == i).astype(int).sum())
    classifiers.append(trained_model)
````

````python
# Convert into vector of tokens that map to our training data[]

diff_cols = [col for col in token_vectors.columns if col not in test_token_vectors.columns]

for col in diff_cols:
    test_token_vectors[col] = 0

test_token_vectors = test_token_vectors[token_vectors.columns]

predictions = np.zeros((2000,1000))
for i in range(len(classifiers)):
    clf = classifiers[i]
    pred = clf.predict_proba(test_token_vectors.values)[:,1]
    predictions[:,i] = pred

predictions    
````

````python
ResNet2 = pd.read_csv('./data/features_test/features_resnet1000_test.csv', header=None)
ResNet2.columns = ['fnum'] + list(range(1000))
ResNet2['fnum'] = ResNet2['fnum'].apply(lambda x: int(x.split('/')[-1].split('.')[0]))
ResNet2.sort_values('fnum', inplace = True)
ResNet2.set_index('fnum', inplace = True)
````

````python
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances

# For each testing vector, find 20 nearest neighbors
similarities = cosine_similarity(ResNet2.values, normalize(predictions))


def reverse(matrix):
    A = np.zeros(matrix.shape)
    for i in range(len(matrix[:,0])):
        row_order_indices = similarities[i].argsort() 
        A[i] = row_order_indices[::-1]
    return A

def similarity_to_results(similarities_matrix, fname):
    #sort_indices = (similarities_matrix*-1).argsort(axis=1)
    sort_indices = similarities.argsort() #reverse(similarities_matrix)
    
    results = {}
    for i in range(2000): # Each test
        top_20 = [
            str(np.where(sort_indices[i,:] == j)[0][0]) for j in range(20)]
        results[str(i)+'.txt'] = '.jpg '.join(top_20) + '.jpg'

    # Create Submission
    results = pd.DataFrame.from_dict(results, orient='index').reset_index()
    results.columns = ['Descritpion_ID','Top_20_Image_IDs']
    results.to_csv(fname, index=False,index_label=False)
    
similarity_to_results(similarities, 'swaggy_submission_euclidean_propersort.csv')
````
